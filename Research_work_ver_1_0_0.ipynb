## Загрузка данных

Get tickers from info file

```python
first_N = 200#10000 # -1
tickers = pd.read_excel(
     os.path.join("./", "outer sources", "Symbols.xlsx"),
     engine='openpyxl',
     sheet_name=1,
)
tickers.columns = list(tickers.iloc[2:,:].iloc[0].values)
tickers = tickers.iloc[3:,:]
tickers_names = 'VZ UNH JNJ PG MRK RTX KO WBA MMM WMT MSFT IBM INTC AAPL CSCO '
tickers_names += 'NKE CAT PFE XOM MCD JPM HD V GS DIS AXP CVX TRV BA ' # DOW
tickers_names += ' '.join(tickers["Ticker"].values[:first_N])
```

```python
first_N = 20000
NASDAQ_names = pd.read_csv(
    os.path.join("./", "outer sources", "Symbols_NASDAQ.csv"),
)["Symbol"].values
tickers_names += ' ' + ' '.join(NASDAQ_names[:first_N:50])
```

```python
# Убираем дубликаты
tickers_names = " ".join(list(set(tickers_names.split())))
tickers_names
```

Load data for this tickers

```python
load_source = ["yfinance", "simfin", "file"][0]

loader = data_loader.DataLoader()
#data = loader.load_data(load_source)
data = loader.load_yfinance(companies=tickers_names, period="1d")
data.head()
```

```python
prices = data['Adj Close'].dropna(axis=1)
prices.head()
```
